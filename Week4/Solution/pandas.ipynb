## Seriese
- One dimensinal array 
- All elements must be the same data type
- Think of it like a column on a table 
- Fixed lengeth 
- Single direction 
- Index Lables if nothing else is specified 

```python
# Series
# Defult index as Label
nums = pd.Series([1,2,3,4])
print(nums)
print(nums[0])

#Create your own lable
names = ["rose", "Ix", "Eric"]
users = pd.Series(names, index = ["C", "H", "H"])
print(users)
print(users["C"])

menu_prices = pd.Series({
    "Coffee": 3.50,
    "Tea": 2.75,
    "Sandwich": 6.25,
    "Salad": 5.50,
    "Cookie": 1.75
})

print(menu_prices["Tea"])
menu_prices
```

## Values and Index
- .value will return the values in a 2d array
- .index retruns lables 

```python
print(menu_prices.values)
print(menu_prices.index)

print("no predefined index lables: ", nums.index)
```

## Update 
- .update is used to update a value in a Series 

```python
#Update One
seattle_temperature = pd.Series([49, 49, None, 50, 54])
seattle_temperature.update(pd.Series([49], index=[2]))

seattle_temperature
```

```python
#Update Multiple 
seattle_temperature = pd.Series([49, 49, None, 50, 54, None])
seattle_temperature.update(pd.Series([49, 55], index=[2,5]))
seattle_temperature
```

## Updaing index lables
- .rename will update an index lable 

```python
seattle_temp = pd.Series(
    [48, 50, 52, 49, 47, 51, 53],
    index=["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
)
seattle_temp_short = seattle_temp.rename({
    "Mon": "M",
    "Tue": "Tu",
    "Wed": "W",
    "Thu": "Th",
    "Fri": "F",
    "Sat": "Sa",
    "Sun": "Su"
})
print("Seattle Temp", seattle_temp)
print("Short Temp", seattle_temp_short)
```

```python
#Inplace 
seattle_temp = pd.Series(
    [48, 50, 52, 49, 47, 51, 53],
    index=["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
)
seattle_temp_short = seattle_temp.rename({
    "Mon": "M",
    "Tue": "Tu",
    "Wed": "W",
    "Thu": "Th",
    "Fri": "F",
    "Sat": "Sa",
    "Sun": "Su"
}, inplace=True)
#Changes the original 
print("Seattle Temp", seattle_temp)
#Does NOT make a copy 
print("Short Temp", seattle_temp_short)
```

## Data Frames 
Data frames are the most used datastructure provided by pandas. 
- Two-dimensional data structure, like a table, with rows representing records and columns representing features.


```python
walkability = pd.DataFrame({
    "City": ["Seattle", "New York", "Los Angeles", "Chicago", "Phoenix"],
    "Walkability": ["High", "Very High", "Medium", "High", "Low"]
})
# Notebooks will even provide a pretty table!
walkability
```

```python
#Other ways to make DataFrames
df = pd.DataFrame([
    {"City": "Seattle", "Walkability": "High"},
    {"City": "New York", "Walkability": "Very High"},
    {"City": "Chicago", "Walkability": "High"}
])
df
```

```python
data = [
    ["Seattle", "High"],
    ["New York", "Very High"],
    ["Chicago", "High"]
]

df = pd.DataFrame(data, columns=["City", "Walkability"])
df
```

```python
data = np.array([
    ["Seattle", "High"],
    ["New York", "Very High"],
    ["Chicago", "High"]
])

df = pd.DataFrame(data, columns=["City", "Walkability"])
df
```

```python
cities = pd.Series(["Seattle", "New York", "Chicago"])
walkability = pd.Series(["High", "Very High", "High"])

df = pd.DataFrame({
    "City": cities,
    "Walkability": walkability
})
df
```

# Indexing 
- DataFrames use indexing in a similar way to Series.

```python

df = pd.DataFrame(
    {
        "Walkability": ["High", "Very High", "High", "Low"],
        "TransitScore": [85, 90, 78, 50],
        "BikeScore": [70, 80, 65, 40]
    })

df.index = ["Seattle", "New York", "Chicago", "Phoenix"]

df
```

```python

df = pd.DataFrame(
    {
        "Walkability": [
            "High", "Very High", "High", "Low",
            "High", "Medium", "High"
        ],
        "TransitScore": [
            85, 90, 78, 50,
            60, 55, 65
        ],
        "BikeScore": [
            70, 80, 65, 40,
            85, 70, 80
        ]
    },
    index=[
        "Seattle", "New York", "Chicago", "Phoenix",
        "Portland", "Boulder", "Madison"
    ]
)

df

# Rename can be used to update index values
df.rename(index={"New York": "NYC"}, inplace=True)

df
```

## Accessing data
- Columns
  - bracket notation df["column"]
  - dot notation df.column
- rows
  - By lable df.loc["index lable"]
  - by position df.iloc[0]

```python
print(df["TransitScore"])
print()
print(df.TransitScore)
print()
print(df.loc["Seattle"])
print(df.iloc[0])
```

```python
# Adding new Columns
df["PopulationMillions"] = [0.75, 8.3, 2.7, 1.6, 0.65, 0.11, 0.27]
df
```

## Performing math on columns
DataFrames allow math operations to be performed on a single or entire columns using other columns.

```python
#Single Column
print(df["TransitScore"].sum())

#Multiple Columns
print(df[["TransitScore", "BikeScore"]].sum())

#Find max
print(df["PopulationMillions"].idxmax())
```

## Operations on entire columns 

```python
df["AvgTransport"] = (df["TransitScore"] + df["BikeScore"]) / 2
print(df["AvgTransport"])
```

```python
bike_per_million = df["BikeScore"] / df["PopulationMillions"]
bike_per_million
```
