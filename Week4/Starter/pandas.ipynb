Pandas are cute bears, and also the name of a powerful Python library for working with datasets, making it an essential tool in any data scientist’s toolkit.

Data scientists spend 60–80% of their time cleaning and organizing data. This may seem excessive until you consider why. If you have bad “garbage” data, your analytics and models will produce bad “garbage” results. Garbage in, garbage out.

Pandas helps us with this by providing tools to fix errors, remove duplicates, and handle inconsistencies, all to improve analysis and model performance.

## Seriese
- One dimensinal array 
- All elements must be the same data type
- Think of it like a column on a table 
- Fixed lengeth 
- Single direction 
- Index Lables if nothing else is specified 

```python
# Series
# Default array index will be out Label

nums = pd.Series([1, 2, 3, 4])
print(nums[0])
nums

#Create your own labels
name = ["Angi", "Momo", "Monkey"]
users = pd.Series(name, index = ["H1", "H2" ,"C"])
print(users["C"])
users



```

```python
menu_price = pd.Series({
    "Coffee": 2.50,
    "Tea": 1.75,
    "Sandwich": 5.00,
    "Salad": 4.25,
    "Cookie": 1.00
})
print(menu_price["Tea"])
menu_price
```

## Values and Index
- .value will return the values in a 2d array
- .index retruns lables 

```python
print("Values: ", menu_price.values)
print()
print("Index: ", menu_price.index)
```

## Update 
- .update is used to update a value in a Series 

```python
seattle_temperature = pd.Series([49, 49, None, 50, 54])
seattle_temperature.update(pd.Series([49], index=[2]))
seattle_temperature
```

```python
seattle_temperature = pd.Series([49, 49, None, 50, 54, None])
seattle_temperature.update(pd.Series([49, 55], index=[2, 5]))
seattle_temperature
```

## Updaing index lables
- .rename will update an index lable 

```python
seattle_temperature = pd.Series(
    [48, 50, 49, 47, 51], index=["Mon", "Tue", "Wed", "Thu", "Fri"]
)
seattle_temp_short = seattle_temperature.rename((
    "Mon : M",
    "Tue: T",
    "Wed: W",
    "Thu: T",
    "Fri: F"
))

print("Short Temp", seattle_temp_short)
print()
print("Seattle Temp: ", seattle_temperature)
```

```python

seattle_temperature = pd.Series(
    [48, 50, 49, 47, 51], index=["Mon", "Tue", "Wed", "Thu", "Fri"]
)
seattle_temp_short = seattle_temperature.rename({
    "Mon: M",
    "Tue: T",
    "Wed: W",
    "Thu: T",
    "Fri: F"
}, inplace=True)

print("Short Temp", seattle_temp_short)
print()
print("Seattle Temp: ", seattle_temperature)
```

## Data Frames 
Data frames are the most used datastructure provided by pandas. 
- Two-dimensional data structure, like a table, with rows representing records and columns representing features.


```python
walkibility = pd.DataFrame({
    "City": ["Seattle", "Portland", "San Francisco", "Chiraq"],
    "Walkibility": ["High", "Very High", "Medium", "Low"]
})
walkibility
```

```python
data = np.array([
    ["Seattle", "High"],
    ["Portland", "Very High"],
    ["San Francisco", "Medium"],
    ["Chiraq", "Low"]
])
df = pd.DataFrame(data, columns=["City", "Walkibility"])
df
```

```python
cities = pd.Series(["Seattle", "Portland", "San Francisco"])
walkibility = pd.Series(["High", "Very High", "Very High"])

df = pd.DataFrame({
    "City": cities,
    "Walkibility": walkibility
})
df
```

# Indexing 
- DataFrames use indexing in a similar way to Series.

```python

df = pd.DataFrame(
    {
        "Walkability": ["High", "Very High", "High", "Low"],
        "TransitScore": [85, 90, 78, 50],
        "BikeScore": [70, 80, 65, 40]
    })

df.index = ["Seattle", "New York", "Chicago", "Phoenix"]

df
```

```python

df = pd.DataFrame(
    {
        "Walkability": [
            "High", "Very High", "High", "Low",
            "High", "Medium", "High"
        ],
        "TransitScore": [
            85, 90, 78, 50,
            60, 55, 65
        ],
        "BikeScore": [
            70, 80, 65, 40,
            85, 70, 80
        ]
    },
    index=[
        "Seattle", "New York", "Chicago", "Phoenix",
        "Portland", "Boulder", "Madison"
    ]
)

df

# Rename can be used to update index values
df.rename(index={"New York": "NYC"}, inplace=True)

df
```

## Accessing data
- Columns
  - bracket notation df["column"]
  - dot notation df.column
- rows
  - By lable df.loc["index lable"]
  - by position df.iloc[0]

```python
print(df["TransitScore"])
print()
print(df.TransitScore)
print()
print(df.loc["Seattle"])
print(df.iloc[0])
```

```python
df["PopulationMillions"] = [0.74, 8.4, 2.7, 1.6, 0.65, 0.11, 0.26]
df
```

## Performing math on columns
DataFrames allow math operations to be performed on a single or entire columns using other columns.

```python
# Single Column
print(df["TransitScore"].sum())

# Multiple Columns
print(df[["TransitScore", "BikeScore"]].sum())
```

## Operations on entire columns 

```python
df["AvgTransport"] = (df["TransitScore"] + df["BikeScore"]) / 2
print(df["AvgTransport"])
```

```python
bike_per_million = df["BikeScore"] / df["PopulationMillions"]
bike_per_million
```
