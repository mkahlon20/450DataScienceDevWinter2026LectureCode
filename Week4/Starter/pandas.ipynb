Pandas are cute bears, and also the name of a powerful Python library for working with datasets, making it an essential tool in any data scientist’s toolkit.

- Cleaning messy datasets 

## Why it matters 
Data scientists spend 60–80% of their time cleaning and organizing data. This may seem excessive until you consider why. If you have bad “garbage” data, your analytics and models will produce bad “garbage” results. Garbage in, garbage out.

Pandas helps us with this by providing tools to fix errors, remove duplicates, and handle inconsistencies, all to improve analysis and model performance.
